### 신문기사 출력
- 시애틀의 대표적인 신문 Seattle Times에서 글을 불러와 파일로 저장하기
- https://www.seattletimes.com/business/real-estate/zillows-zestimate-overvalued-a-washington-home-by-700-percent-in-a-case-of-algorithms-gone-wrong/

```python
# 라이브러리 import
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# 불러오려는 url
url = 'https://www.seattletimes.com/business/what-ban-king-countys-fireworks-rules-havent-discouraged-these-generations-old-tribal-vendors/'

# urlopen
web = urlopen(url)

# BeautifulSoup 으로 html 구조 파싱
web_page = BeautifulSoup(web, 'html.parser')
```

```python
web_page
```

```python
# Settle Times 기사 불러오기

with open('tims_txt1.txt', 'w', encoding = 'utf-8') as f:
    times = web_page.find('div', {'class':'article-content'})
    article = times.find_all('p')
    
    for contents in article:
        print(contents.get_text())
        f.write(contents.get_text() + '\n')
```

### Brunch 게시글 출력하고 저장하기
- 브런치에 실린 게시글을 불러와 파일로 저장하기
- https://brunch.co.kr/

```python
# 불러오려는 url 저장
url = 'https://brunch.co.kr/@clinemr85/216'

# urlopen
web = urlopen(url)

# BeautifulSoup으로 web페이지 html 구조 파싱
web_page = BeautifulSoup(web, 'html.parser')
```

```python
web_page
```

```python
# Brunch 글 불러오기
with open('brunch1.txt','w', encoding='utf-8') as f:
    brunch_all = web_page.find('div', {'class':'wrap_body'})
    brunch_txt = brunch_all.find_all('p')
    
    for brunch2 in brunch_txt:
        print(brunch2.get_text())
        f.write(brunch2.get_text() + '\n')
```

```python
# Save multiple articles automatically

# @imagineer 님의 글 10개를 자동으로 가져와 저장하는 코드를 작성해보세요!
# Hint 1 : str에는 str만 더할 수 있어요!
# Hint 2 : @imagineer님 글의 URL은 몇 번부터 시작할까요?

# 에러나는 url 담을곳
error_url = []

# 100번돌린다
for i in range(50):
    try:
        # url 파라미터 변수 지정
        url = 'https://brunch.co.kr/@clinemr85/' + str(i)
        web = urlopen(url)
        web_page = BeautifulSoup(web, 'html.parser')
        
        # 저장할 텍스트 지정
        with open('test_all.txt', 'a', encoding='utf-8') as f:
            
            # 담을 내용
            all_text = web_page.find('div', {'class':'cover_item'})
            article1 = all_text.find_all('h1')  
            article2 = all_text.find_all('p')
            article3 = all_text.find_all('span')
            all_article = article1 + article2 + article3
            
            # 반복문 사용하여 담을 내용을 저장
            for contents in all_article:
                print(contents.get_text())
                f.write(contents.get_text() + '\n')
    except:
        # 에러난 url은 변수에 따로 모아놓는다
        error_url.append(url)
        
```
