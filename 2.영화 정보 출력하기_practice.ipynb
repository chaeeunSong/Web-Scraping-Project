### 영화 정보 출력하기
네이버영화 사이트에서 각종 영화 정보를 스크래이핑

```python
# 필요한 라이브러리 불러오기
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
# 불러오려는 url 입력하기
url = 'https://movie.naver.com/movie/bi/mi/basic.naver?code=198413'

# urlopen 함수를 통해 web 변수를 생성
web = urlopen(url)

# BeautifulSoup으로 web 페이지상의 html 구조를 파싱
web_page = BeautifulSoup(web, 'html.parser')
```

```python
# 영화 제목을 출력한다
movie_title = web_page.find('h3', {'class' : 'h_movie'}).find('a')

print(movie_title.get_text())
```

```python
# 영화 줄거리를 출력한다
movie_article = web_page.find('p', {'class':'con_tx'})

print(movie_article.get_text())
```

```python
# 감독 이름 & 출연 배우 이름 스크래이핑

# 불러오려는 url
url = 'https://movie.naver.com/movie/bi/mi/detail.naver?code=198413'

# urlopen
web = urlopen(url)

# BeautifulSoup으로 html 파싱
web_page = BeautifulSoup(web, 'html.parser')
```

```python
web_page
```

```python
# 감독 이름을 출력
director_name = web_page.find('div', {'class' : 'dir_product'}).find('a')

print(director_name.get_text())
```

```python
# 출연 배우 이름 출력
actor_all = web_page.find('div', {'class' : 'lst_people_area'})

actor_name = actor_all.find_all('a', {'class':'k_name'})

for actor_list in actor_name:
    print(actor_list.get_text())
```

```python
# 영화 리뷰 스크레이핑
# 불러오려는 url 입력
url = 'https://movie.naver.com/movie/bi/mi/review.naver?code=198413'

# urlopen
web = urlopen(url)

# BeautifulSoup으로 html 구조 파싱
web_page = BeautifulSoup(web, 'html.parser')
```

```python
# 영화 리뷰 제목 출력
review_all = web_page.find('ul', {'class':'rvw_list_area'})
review_title = review_all.find_all('strong')

# 영화 리뷰 텍스트 출력
review_article = review_all.find_all('p')

for review_text in review_title:
    print(review_text.get_text())

for reveiw_text2 in review_article:
    print(reveiw_text2.get_text())
```
