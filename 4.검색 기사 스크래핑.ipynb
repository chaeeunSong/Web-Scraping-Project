```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EA%B8%88%EB%A6%AC%EC%9D%B8%EC%83%81&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=27&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1'
web = urlopen(url)
web_page = BeautifulSoup(web, 'html.parser')
```

```python
wrap = web_page.find('div',{'class':'group_news'})
titles = wrap.find_all('a',{'class':'news_tit'})

for title in titles:
    print(title.get_text())
```

```python
error_urls = []

for i in range(20):
    try:
        url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EA%B8%88%EB%A6%AC%EC%9D%B8%EC%83%81&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=27&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=' + str(i)
        web = urlopen(url)
        source = BeautifulSoup(web, 'html.parser')
        
        with open('search_news_title.txt', 'a', encoding='utf-8') as f:
            
            #all_box = source.find('li', {'class':'bx'})
            all_title = source.find_all('a',{'class':'news_tit'})
            
            for content in all_title:
                print(content.get_text())
                f.write(content.get_text() + '\n')
    except:
            error_urls.append(url)
```
